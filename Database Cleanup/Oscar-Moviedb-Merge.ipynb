```python
#Add Dependencies
import pandas as pd
import numpy as np
```

```python
#Setup File System
read_oscars_file = "../Resources/oscars_clean.csv"
read_movies_file = "../Resources/movies_clean.csv"
write_file = "../Resources/movies_oscars_combined.csv"

#read_file = "/Users/archieanand/Desktop/Analysis/Working/the_oscar_award.csv"
#write_file = "/Users/archieanand/Desktop/Analysis/Working/oscars_clean.csv"
```

```python
# Import our Oscar dataset
oscars_df = pd.read_csv(read_oscars_file)
oscars_df.head()
```

```python
oscars_df.info()
```

```python
oscars_df.sort_values(by='winner', ascending=False)
```

```python
oscars_df = oscars_df.drop_duplicates(subset='film', keep='first')
```

```python
movies_df = pd.read_csv(read_movies_file)
movies_df.head()
```

```python
movies_df.info()
```

```python
combined_df = pd.merge(movies_df, oscars_df, how='left', left_on='title', right_on='film' )
combined_df.shape
```

```python
combined_df = combined_df.drop_duplicates()
```
